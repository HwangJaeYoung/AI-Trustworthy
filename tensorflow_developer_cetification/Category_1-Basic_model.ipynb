```python
# 필요한 패키지 import
import numpy as np

# STEP 1: 필요한 모듈 import
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
import tensorflow as tf
import warnings
warnings.filterwarnings('ignore')
```

```python
# STEP 2: 데이터 전처리
xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([5.0, 6.0, 7.0, 8.0, 9.0, 10.0], dtype=float)
```

```python
# STEP 3: 모델의 정의 (modeling)
# list: input_shape=[1] / tuple: input_shape=(1, ) 가능
# int type: input_shape=(1) 불가능
# 1번 문제 같은 경우는 xs가 한개만 들어가기 때문에 input_shape=[1]

# model = Sequential([
#     Dense(3, input_shape=[1]), #input_shape는 list나 tuple이 와야한다.
#     Dense(4),
#     Dense(4),
#     Dense(1),
# ])
```

```python
# 하지만, 본 문제를 푸는데, 데이터의 복잡도가 단순하기 때문에, 아래와같이 Dense의 unit은 1로 지정
# input_shape=[1], input_shape=(1,)
model = Sequential([
    Dense(1, input_shape=[1]),
])
```

```python
# STEP 4: 모델의 생성 (compile)
model.compile(optimizer='sgd', loss='mse')
```

```python
# STEP 5: 학습 (fit)
model.fit(xs, ys, epochs=1200, verbose=0) #verbose=1: 모든 로그 출력, verbose=0: 로그 출력 중지
```

```python
# output
# 16.000046
model.predict([10.0])
```
